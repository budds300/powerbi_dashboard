```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from scipy import stats

# Load the CSV file
# df = pd.read_csv('total_transactions_2019.csv')
df['date'] = pd.to_datetime(df['date'])

# Create a daily sales dataframe
daily_sales = df.groupby('date')['order_total'].sum().reset_index()
daily_sales.set_index('date', inplace=True)

# Perform seasonal decomposition
result = seasonal_decompose(daily_sales['order_total'], model='additive', period=7)

# Plot the decomposition
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 20))
result.observed.plot(ax=ax1)
ax1.set_title('Observed')
result.trend.plot(ax=ax2)
ax2.set_title('Trend')
result.seasonal.plot(ax=ax3)
ax3.set_title('Seasonal')
result.resid.plot(ax=ax4)
ax4.set_title('Residual')
plt.tight_layout()
plt.show()

# Create a box plot for monthly sales
monthly_sales = df.groupby(['month_name', 'month'])['order_total'].sum().reset_index()
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_sales['month_name'] = pd.Categorical(monthly_sales['month_name'], categories=month_order, ordered=True)
monthly_sales = monthly_sales.sort_values('month_name')

plt.figure(figsize=(15, 6))
sns.boxplot(x='month_name', y='order_total', data=monthly_sales)
plt.title('Monthly Sales Distribution')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

# Create a line plot with a 7-day moving average
daily_sales['7_day_ma'] = daily_sales['order_total'].rolling(window=7).mean()

plt.figure(figsize=(15, 6))
plt.plot(daily_sales.index, daily_sales['order_total'], label='Daily Sales')
plt.plot(daily_sales.index, daily_sales['7_day_ma'], label='7-day Moving Average', color='red')
plt.title('Daily Sales with 7-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Perform and plot autocorrelation
from statsmodels.graphics.tsaplots import plot_acf

plt.figure(figsize=(15, 6))
plot_acf(daily_sales['order_total'], lags=50)
plt.title('Autocorrelation of Daily Sales')
plt.show()

print("Visualizations have been created and displayed.")

# Calculate correlation between day of week and sales
df['day_of_week'] = df['date'].dt.dayofweek
day_sales = df.groupby('day_of_week')['order_total'].sum().reset_index()
correlation, p_value = stats.pearsonr(day_sales['day_of_week'], day_sales['order_total'])

print(f"\
Correlation between day of week and sales: {correlation:.4f}")
print(f"P-value: {p_value:.4f}")

# Calculate the coefficient of variation for each month
monthly_cv = df.groupby('month_name')['order_total'].agg(['mean', 'std']).reset_index()
monthly_cv['cv'] = monthly_cv['std'] / monthly_cv['mean']
monthly_cv = monthly_cv.sort_values('cv', ascending=False)

print("\
Coefficient of Variation for each month (higher values indicate more variability):")
print(monthly_cv[['month_name', 'cv']])
```

```python
# How do sales compare to previous months or weeks? 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
df = pd.read_csv('total_transactions_2019.csv')
df['date'] = pd.to_datetime(df['date'])

# Group sales by month
monthly_sales = df.groupby(df['date'].dt.to_period('M'))['order_total'].sum().reset_index()
monthly_sales['date'] = monthly_sales['date'].dt.to_timestamp()

# Calculate month-over-month growth
monthly_sales['MoM_Growth'] = monthly_sales['order_total'].pct_change()

# Plot monthly sales trend
plt.figure(figsize=(14, 6))
plt.plot(monthly_sales['date'], monthly_sales['order_total'], marker='o')
plt.title('Monthly Sales Trend (2019)')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Display monthly sales summary
print("Monthly Sales Summary:")
print(monthly_sales)

# Calculate and display average monthly sales
avg_monthly_sales = monthly_sales['order_total'].mean()
print(f"\
Average Monthly Sales: ${avg_monthly_sales:.2f}")

# Identify best and worst performing months
best_month = monthly_sales.loc[monthly_sales['order_total'].idxmax()]
worst_month = monthly_sales.loc[monthly_sales['order_total'].idxmin()]

print(f"\
Best performing month: {best_month['date'].strftime('%B %Y')} with ${best_month['order_total']:.2f} in sales")
print(f"Worst performing month: {worst_month['date'].strftime('%B %Y')} with ${worst_month['order_total']:.2f} in sales")
```

```python
# Hypothesis VI
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the data
# df = pd.read_csv('total_transactions_2019.csv')

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])

# 1. Analyze sales trends over time
daily_sales = df.groupby('date')['order_total'].sum().reset_index()

# 2. Analyze product category performance
category_performance = df.groupby('product_category')['order_total'].agg(['sum', 'count', 'mean']).reset_index()
category_performance = category_performance.sort_values('sum', ascending=False)

# 3. Analyze customer purchase frequency
purchase_frequency = df.groupby('order_id')['quantity_ordered'].sum().value_counts().sort_index()

# 4. Analyze time of day sales patterns
df['hour'] = pd.to_datetime(df['time']).dt.hour
hourly_sales = df.groupby('hour')['order_total'].sum().reset_index()

# Visualizations
fig, axs = plt.subplots(2, 2, figsize=(20, 20))

# Sales trend
axs[0, 0].plot(daily_sales['date'], daily_sales['order_total'])
axs[0, 0].set_title('Daily Sales Trend')
axs[0, 0].set_xlabel('Date')
axs[0, 0].set_ylabel('Total Sales')

# Category performance
axs[0, 1].bar(category_performance['product_category'], category_performance['sum'])
axs[0, 1].set_title('Sales by Product Category')
axs[0, 1].set_xlabel('Product Category')
axs[0, 1].set_ylabel('Total Sales')
axs[0, 1].tick_params(axis='x', rotation=45)

# Purchase frequency
axs[1, 0].bar(purchase_frequency.index, purchase_frequency.values)
axs[1, 0].set_title('Purchase Frequency Distribution')
axs[1, 0].set_xlabel('Number of Items per Order')
axs[1, 0].set_ylabel('Frequency')

# Hourly sales pattern
axs[1, 1].bar(hourly_sales['hour'], hourly_sales['order_total'])
axs[1, 1].set_title('Sales by Hour of Day')
axs[1, 1].set_xlabel('Hour')
axs[1, 1].set_ylabel('Total Sales')

plt.tight_layout()
plt.show()

# Print summary statistics
print("Category Performance Summary:")
print(category_performance)

print("\
Purchase Frequency Summary:")
print(purchase_frequency.head())

print("\
Hourly Sales Summary:")
print(hourly_sales)

# Conduct statistical tests
# 1. Test for trend in daily sales
trend_test = stats.linregress(range(len(daily_sales)), daily_sales['order_total'])
print("\
Trend Test (Daily Sales):")
print(f"p-value: {trend_test.pvalue}")

# 2. Test for differences in category performance
category_sales = [group['order_total'].values for name, group in df.groupby('product_category')]
category_test = stats.f_oneway(*category_sales)
print("\
Category Performance Test:")
print(f"p-value: {category_test.pvalue}")

# 3. Test for uniformity in hourly sales
hourly_test = stats.chisquare(hourly_sales['order_total'])
print("\
Hourly Sales Uniformity Test:")
print(f"p-value: {hourly_test.pvalue}")
```
