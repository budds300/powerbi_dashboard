```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose
from scipy import stats

# Load the CSV file
# df = pd.read_csv('total_transactions_2019.csv')
df['date'] = pd.to_datetime(df['date'])

# Create a daily sales dataframe
daily_sales = df.groupby('date')['order_total'].sum().reset_index()
daily_sales.set_index('date', inplace=True)

# Perform seasonal decomposition
result = seasonal_decompose(daily_sales['order_total'], model='additive', period=7)

# Plot the decomposition
fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(15, 20))
result.observed.plot(ax=ax1)
ax1.set_title('Observed')
result.trend.plot(ax=ax2)
ax2.set_title('Trend')
result.seasonal.plot(ax=ax3)
ax3.set_title('Seasonal')
result.resid.plot(ax=ax4)
ax4.set_title('Residual')
plt.tight_layout()
plt.show()

# Create a box plot for monthly sales
monthly_sales = df.groupby(['month_name'])['order_total'].sum().reset_index()
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']
monthly_sales['month_name'] = pd.Categorical(monthly_sales['month_name'], categories=month_order, ordered=True)
monthly_sales = monthly_sales.sort_values('month_name')

plt.figure(figsize=(15, 6))
sns.boxplot(x='month_name', y='order_total', data=monthly_sales)
plt.title('Monthly Sales Distribution')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()

# Create a line plot with a 7-day moving average
daily_sales['7_day_ma'] = daily_sales['order_total'].rolling(window=7).mean()

plt.figure(figsize=(15, 6))
plt.plot(daily_sales.index, daily_sales['order_total'], label='Daily Sales')
plt.plot(daily_sales.index, daily_sales['7_day_ma'], label='7-day Moving Average', color='red')
plt.title('Daily Sales with 7-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()

# Perform and plot autocorrelation
from statsmodels.graphics.tsaplots import plot_acf

plt.figure(figsize=(15, 6))
plot_acf(daily_sales['order_total'], lags=50)
plt.title('Autocorrelation of Daily Sales')
plt.show()

print("Visualizations have been created and displayed.")

# Calculate correlation between day of week and sales
df['day_of_week'] = df['date'].dt.dayofweek
day_sales = df.groupby('day_of_week')['order_total'].sum().reset_index()
correlation, p_value = stats.pearsonr(day_sales['day_of_week'], day_sales['order_total'])

print(f"\
Correlation between day of week and sales: {correlation:.4f}")
print(f"P-value: {p_value:.4f}")

# Calculate the coefficient of variation for each month
monthly_cv = df.groupby('month_name')['order_total'].agg(['mean', 'std']).reset_index()
monthly_cv['cv'] = monthly_cv['std'] / monthly_cv['mean']
monthly_cv = monthly_cv.sort_values('cv', ascending=False)

print("\
Coefficient of Variation for each month (higher values indicate more variability):")
print(monthly_cv[['month_name', 'cv']])
```

```python
# Hypothesis Testing V
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_csv('total_transactions_2019.csv')

# Calculate Z-scores for 'order_total'
z_scores = np.abs(stats.zscore(df['order_total']))

# Identify outliers using Z-score method (threshold = 3)
z_outliers = df[z_scores > 3]

# Calculate IQR for 'order_total'
Q1 = df['order_total'].quantile(0.25)
Q3 = df['order_total'].quantile(0.75)
IQR = Q3 - Q1

# Identify outliers using IQR method
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
iqr_outliers = df[(df['order_total'] < lower_bound) | (df['order_total'] > upper_bound)]

# Print the number of outliers detected by each method
print("Number of outliers detected by Z-score method:", len(z_outliers))
print("Number of outliers detected by IQR method:", len(iqr_outliers))

# Visualize the distribution of 'order_total' with outliers
plt.figure(figsize=(12, 6))
sns.boxplot(x=df['order_total'])
plt.title('Boxplot of Order Total with Outliers')
plt.xlabel('Order Total')
plt.show()
```

Z-score Method:
The Z-score method identified 2,246 outliers in the 'order_total' column. This method considers data points that are more than 3 standard deviations away from the mean as outliers.

IQR (Interquartile Range) Method:
The IQR method identified 17,808 outliers in the 'order_total' column. This method considers data points below Q1 - 1.5IQR or above Q3 + 1.5IQR as outliers, where Q1 is the 25th percentile and Q3 is the 75th percentile.

Boxplot Visualization:
The boxplot clearly shows the presence of numerous outliers in the 'order_total' column. The long whiskers and many points beyond the whiskers indicate a significant number of extreme values.

Analysis:

Difference in Outlier Count: The IQR method identified considerably more outliers than the Z-score method. This difference is due to the different assumptions and sensitivities of these methods:

The Z-score method assumes a normal distribution and is less sensitive to extreme outliers.
The IQR method is non-parametric and more sensitive to skewed distributions.
